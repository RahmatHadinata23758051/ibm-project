```python
# === Tahap 3: Data Preprocessing & Merging (1 cell) ===
import os
import re
import pandas as pd

# ------------------------------------------------------
# 1) Utility: fungsi normalisasi nama negara (lowercase, trim, hilangkan teks dalam kurung, dll.)
# ------------------------------------------------------
def norm_country(s: str) -> str:
    if pd.isna(s):
        return s
    x = s.strip().lower()
    # hilangkan teks di dalam kurung, mis. "iran (islamic republic of)" -> "iran"
    x = re.sub(r"\s*\(.*?\)\s*", "", x)
    # ganti & dengan 'and', hapus karakter non-alfabet (kecuali spasi)
    x = x.replace("&", " and ")
    x = re.sub(r"[^a-z\s\-]", "", x)
    x = re.sub(r"\s+", " ", x).strip()
    # mapping beberapa varian umum -> bentuk standar
    mapping = {
        "united states of america": "united states",
        "russian federation": "russia",
        "viet nam": "vietnam",
        "korea republic of": "south korea",
        "republic of korea": "south korea",
        "iran islamic republic of": "iran",
        "venezuela bolivarian republic of": "venezuela",
        "egypt arab republic": "egypt",
        "kyrgyz republic": "kyrgyzstan",
        "lao pdr": "laos",
        "yemen rep": "yemen",
        "congo dem rep": "congo dem republic",
        "congo rep": "congo republic",
        "gambia the": "gambia",
        "brunei darussalam": "brunei",
        "bahamas the": "bahamas",
        "cabo verde": "cape verde",
        "cote divoire": "cote d ivoire",
        "st lucia": "saint lucia",
        "st martin french part": "saint martin french part",
        "st vincent and the grenadines": "saint vincent and the grenadines",
        "sao tome and principe": "sao tome and principe",
        "virgin islands u.s.": "virgin islands us",
        "virgin islands british": "virgin islands british",
        "west bank and gaza": "west bank and gaza",
        "slovak republic": "slovakia",
        "czech republic": "czechia"
    }
    return mapping.get(x, x)


# ------------------------------------------------------
# 2) Standardisasi nama negara di kedua dataset
# ------------------------------------------------------
print("Standardizing country names...")
df_co2['country_name_norm'] = df_co2['country_name'].apply(norm_country)
df_pollution['Country_norm'] = df_pollution['Country'].apply(norm_country)

print("\nSample normalized country names (df_co2):")
display(df_co2['country_name_norm'].sample(10).unique())
print("\nSample normalized country names (df_pollution):")
display(df_pollution['Country_norm'].sample(10).unique())

# ------------------------------------------------------
# 3) Agregasi df_pollution ke level negara (mean per negara)
# ------------------------------------------------------
print("\nAggregating df_pollution by country...")
# Pilih kolom numerik utama yang relevan untuk AQI
pollution_cols = ['AQI Value', 'CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']
df_pollution_agg = df_pollution.groupby('Country_norm')[pollution_cols].mean().reset_index()
df_pollution_agg.rename(columns={'Country_norm': 'country_name_norm'}, inplace=True) # Rename for merging

print("\nAggregated df_pollution (first 5 rows):")
display(df_pollution_agg.head())
print("\nShape of aggregated df_pollution:", df_pollution_agg.shape)


# ------------------------------------------------------
# 4) Ambil level emisi CO terbaru yang representatif (rata-rata 2017-2019, simpan 2019)
# ------------------------------------------------------
print("\nProcessing df_co2 for recent years...")
df_co2_recent = df_co2[df_co2['year'].isin([2017, 2018, 2019])].copy()

# Hitung rata-rata CO2 2017-2019 per negara
df_co2_avg_recent = df_co2_recent.groupby('country_name_norm')['value'].mean().reset_index()
df_co2_avg_recent.rename(columns={'value': 'co2_emission_kt_avg_2017_2019'}, inplace=True)

# Ambil nilai CO2 tahun 2019 per negara
df_co2_2019 = df_co2[df_co2['year'] == 2019].copy()
df_co2_2019 = df_co2_2019[['country_name_norm', 'value']]
df_co2_2019.rename(columns={'value': 'co2_emission_kt_2019'}, inplace=True)

# Gabungkan rata-rata dan nilai 2019
df_co2_processed = pd.merge(df_co2_avg_recent, df_co2_2019, on='country_name_norm', how='left')


print("\nProcessed df_co2 (first 5 rows):")
display(df_co2_processed.head())
print("\nShape of processed df_co2:", df_co2_processed.shape)


# ------------------------------------------------------
# 5) Merge di level negara
# ------------------------------------------------------
print("\nMerging datasets...")
df_merged = pd.merge(df_pollution_agg, df_co2_processed, on='country_name_norm', how='inner')

print("\nMerged dataframe (first 5 rows):")
display(df_merged.head())
print("\nShape of merged dataframe:", df_merged.shape)

# ------------------------------------------------------
# 6) Output & Verifikasi
# ------------------------------------------------------
print("\n=== Merge Summary ===")
print(f"Jumlah negara unik di df_pollution_agg: {df_pollution_agg['country_name_norm'].nunique()}")
print(f"Jumlah negara unik di df_co2_processed: {df_co2_processed['country_name_norm'].nunique()}")
print(f"Jumlah negara match (setelah inner merge): {df_merged['country_name_norm'].nunique()}")

# Hitung jumlah negara yang tidak match (ada di salah satu tapi tidak di yang lain)
all_countries_pollution = set(df_pollution_agg['country_name_norm'])
all_countries_co2 = set(df_co2_processed['country_name_norm'])
unmatched_in_pollution = all_countries_pollution - all_countries_co2
unmatched_in_co2 = all_countries_co2 - all_countries_pollution

print(f"Jumlah negara di df_pollution_agg yang tidak ada di df_co2_processed: {len(unmatched_in_pollution)}")
print(f"Jumlah negara di df_co2_processed yang tidak ada di df_pollution_agg: {len(unmatched_in_co2)}")

print("\nMerged dataframe info:")
df_merged.info()

# ------------------------------------------------------
# 7) Simpan hasil merge ke CSV
# ------------------------------------------------------
print("\nSaving merged dataframe to CSV...")

# Path pertama
output_path_colab = '/content/combined_country_level.csv'
df_merged.to_csv(output_path_colab, index=False)
print(f"Saved to {output_path_colab}")

# Path kedua (buat folder jika belum ada)
output_dir_ibm = '/content/ibm-project/data/processed/'
os.makedirs(output_dir_ibm, exist_ok=True)
output_path_ibm = os.path.join(output_dir_ibm, 'combined_country_level.csv')
df_merged.to_csv(output_path_ibm, index=False)
print(f"Saved to {output_path_ibm}")

print("\nTahap 3 (Data Preprocessing & Merging) Selesai.")
```

```python
# === Tahap 5: Advanced Analysis ===

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from statsmodels.tsa.arima.model import ARIMA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram, linkage
import plotly.express as px
import os
import warnings
import re

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')

print("=== Starting Advanced Analysis ===")

# ----------------------------------------------------------------------------------------------------
# 1) Data Preparation for Advanced Analysis
# ----------------------------------------------------------------------------------------------------
print("\n1. Data Preparation...")

# Load Merged Data (assuming it was saved in the previous step)
merged_data_path = '/content/ibm-project/data/combined_country_level.csv'
if not os.path.exists(merged_data_path):
    print(f"Error: Merged data file not found at {merged_data_path}. Please run previous steps.")
    # Create an empty DataFrame to prevent errors in subsequent steps
    df_merged = pd.DataFrame()
else:
    df_merged = pd.read_csv(merged_data_path)
    print("Merged data loaded successfully.")
    # Drop rows with NaN values in key columns if any remained after merging
    df_merged.dropna(subset=['co2_emission_kt_avg_2017_2019', 'AQI Value'], inplace=True)
    print(f"Shape of merged data after dropping NaNs: {df_merged.shape}")
    # Ensure 'country_name_norm' is treated as string
    df_merged['country_name_norm'] = df_merged['country_name_norm'].astype(str)


# For Forecasting, we need the original time-series CO2 data.
# Assuming df_co2 from previous steps is available in the environment.
# If not, you would need to reload it or pass it.
# For this combined cell, let's assume it's available or reload if needed.
try:
    df_co2 = pd.read_csv("/content/ibm-project/ibm/co2_emissions_kt_by_country.csv")
    df_co2['country_name'] = df_co2['country_name'].str.strip().str.lower()
    # Use the same normalization function as before
    def norm_country(s: str) -> str:
        if pd.isna(s):
            return s
        x = s.strip().lower()
        x = re.sub(r"\s*\(.*?\)\s*", "", x)
        x = x.replace("&", " and ")
        x = re.sub(r"[^a-z\s\-]", "", x)
        x = re.sub(r"\s+", " ", x).strip()
        mapping = {
            "united states of america": "united states",
            "russian federation": "russia",
            "viet nam": "vietnam",
            "korea republic of": "south korea",
            "republic of korea": "south korea",
            "iran islamic republic of": "iran",
            "venezuela bolivarian republic of": "venezuela",
            "egypt arab republic": "egypt",
            "kyrgyz republic": "kyrgyzstan",
            "lao pdr": "laos",
            "yemen rep": "yemen",
            "congo dem rep": "congo dem republic",
            "congo rep": "congo republic",
            "gambia the": "gambia",
            "brunei darussalam": "brunei",
            "bahamas the": "bahamas",
            "cabo verde": "cape verde",
            "cote divoire": "cote d ivoire",
            "st lucia": "saint lucia",
            "st martin french part": "saint martin french part",
            "st vincent and the grenadines": "saint vincent and the grenadines",
            "sao tome and principe": "sao tome and principe",
            "virgin islands u.s.": "virgin islands us",
            "virgin islands british": "virgin islands british",
            "west bank and gaza": "west bank and gaza",
            "slovak republic": "slovakia",
            "czech republic": "czechia"
        }
        return mapping.get(x, x)

    df_co2['country_name_norm'] = df_co2['country_name'].apply(norm_country)

    # Aggregate global CO2 emissions per year for global forecasting
    df_co2_global = df_co2.groupby('year')['value'].sum().reset_index()
    df_co2_global.set_index('year', inplace=True)
    print("\nGlobal CO2 data for forecasting:")
    display(df_co2_global.head())

    # Note: For AQI forecasting, the current pollution dataset is a snapshot, not time series.
    # We cannot perform time series forecasting on AQI with this dataset structure.
    # The forecasting section will focus on CO2 emissions.
    print("\nNote: AQI forecasting is not possible with the current dataset structure (snapshot). Focusing on CO2 forecasting.")

except FileNotFoundError:
    print("\nWarning: Original df_co2 not found. CO2 forecasting cannot be performed.")
    df_co2_global = None


# Data for Clustering: Use df_merged (already loaded and cleaned)
if df_merged.empty:
    print("\nSkipping Clustering: Merged data not available.")
else:
    clustering_data = df_merged[['co2_emission_kt_avg_2017_2019', 'AQI Value']].copy()
    print("\nData for Clustering (CO2 Avg 2017-2019 vs AQI Value):")
    display(clustering_data.head())

# ----------------------------------------------------------------------------------------------------
# 2) Forecasting (Linear Regression) - CO2 Global
# ----------------------------------------------------------------------------------------------------
if df_co2_global is not None:
    print("\n2. Forecasting (Linear Regression) for Global CO2 Emissions...")
    # Prepare data for Linear Regression
    # X is year (reshaped), y is CO2 value
    X = df_co2_global.index.values.reshape(-1, 1)
    y = df_co2_global['value'].values

    # Train Linear Regression model
    lin_reg = LinearRegression()
    lin_reg.fit(X, y)

    # Predict on historical data
    y_pred_hist = lin_reg.predict(X)

    # Forecast for next 5 years (2020-2024)
    future_years = np.arange(df_co2_global.index.max() + 1, df_co2_global.index.max() + 6).reshape(-1, 1)
    y_forecast_lin_reg = lin_reg.predict(future_years)

    # Evaluate Model (MSE)
    from sklearn.metrics import mean_squared_error
    mse_lin_reg = mean_squared_error(y, y_pred_hist)
    rmse_lin_reg = np.sqrt(mse_lin_reg)
    print(f"Linear Regression MSE: {mse_lin_reg:.2f}")
    print(f"Linear Regression RMSE: {rmse_lin_reg:.2f}")


    # Visualize Linear Regression Forecast
    plt.figure(figsize=(12, 6))
    plt.plot(df_co2_global.index, y, label='Historical Data')
    plt.plot(df_co2_global.index, y_pred_hist, label='Linear Regression Fit', linestyle='--')
    plt.plot(future_years.flatten(), y_forecast_lin_reg, label='Linear Regression Forecast (2020-2024)', linestyle='--')
    plt.title('Global CO2 Emissions Forecast using Linear Regression')
    plt.xlabel('Year')
    plt.ylabel('CO2 Emissions (kt)')
    plt.legend()
    plt.grid(True)
    plt.savefig("co2_forecast_linear_regression.png", dpi=300, bbox_inches="tight")
    plt.show()
    print("Linear Regression CO2 forecast plot saved as co2_forecast_linear_regression.png")

    # Display Forecasted values
    print("\nLinear Regression Forecast (2020-2024):")
    forecast_df_lin_reg = pd.DataFrame({'Year': future_years.flatten(), 'Forecasted CO2 (kt)': y_forecast_lin_reg})
    display(forecast_df_lin_reg)

else:
    print("\nSkipping Linear Regression Forecasting: Global CO2 data not available.")


# ----------------------------------------------------------------------------------------------------
# 3) Forecasting (ARIMA) - CO2 Global
# ----------------------------------------------------------------------------------------------------
if df_co2_global is not None:
    print("\n3. Forecasting (ARIMA) for Global CO2 Emissions...")
    # ARIMA requires the data to be stationary (mean, variance, and seasonality are constant over time).
    # A common approach is to difference the data. Let's try a simple ARIMA(5,1,0) - p=5, d=1, q=0.
    # d=1 means first-order differencing to make the series stationary.
    # For simplicity in this combined cell, we won't perform detailed model order selection (ACF/PACF).

    try:
        # Fit ARIMA model
        # order=(p, d, q)
        arima_model = ARIMA(df_co2_global['value'], order=(5, 1, 0)) # Example order
        arima_result = arima_model.fit()

        print(arima_result.summary()) # Display model summary

        # Plot residuals for diagnostics
        print("\nARIMA Residuals Plot:")
        arima_result.resid.plot()
        plt.title('ARIMA Model Residuals')
        plt.savefig("arima_residuals_plot.png", dpi=300, bbox_inches="tight")
        plt.show()
        print("ARIMA residuals plot saved as arima_residuals_plot.png")


        # Forecast for the next 5 years (adjust steps based on your data's frequency - here assumes yearly)
        n_forecast_steps = 5 # Forecast for 5 years
        arima_forecast = arima_result.forecast(steps=n_forecast_steps)

        # Create future index for forecasting
        future_index_arima = pd.RangeIndex(start=df_co2_global.index.max() + 1, stop=df_co2_global.index.max() + 1 + n_forecast_steps)
        arima_forecast.index = future_index_arima


        # Visualize ARIMA Forecast
        plt.figure(figsize=(12, 6))
        plt.plot(df_co2_global.index, df_co2_global['value'], label='Historical Data')
        plt.plot(arima_forecast.index, arima_forecast, label='ARIMA Forecast (2020-2024)', linestyle='--')
        plt.title('Global CO2 Emissions Forecast using ARIMA')
        plt.xlabel('Year')
        plt.ylabel('CO2 Emissions (kt)')
        plt.legend()
        plt.grid(True)
        plt.savefig("co2_forecast_arima.png", dpi=300, bbox_inches="tight")
        plt.show()
        print("ARIMA CO2 forecast plot saved as co2_forecast_arima.png")

        # Display Forecasted values
        print("\nARIMA Forecast (2020-2024):")
        forecast_df_arima = pd.DataFrame({'Year': arima_forecast.index, 'Forecasted CO2 (kt)': arima_forecast.values})
        display(forecast_df_arima)

    except Exception as e:
        print(f"\nError during ARIMA forecasting: {e}")
        print("Skipping ARIMA forecasting.")

else:
    print("\nSkipping ARIMA Forecasting: Global CO2 data not available.")


# ----------------------------------------------------------------------------------------------------
# 4) Clustering (K-Means) - Countries by CO2 Avg and AQI Value
# ----------------------------------------------------------------------------------------------------
if 'clustering_data' in locals() and not clustering_data.empty:
    print("\n4. Clustering (K-Means) Countries by CO2 and AQI...")

    # Select features for clustering
    X_cluster = clustering_data[['co2_emission_kt_avg_2017_2019', 'AQI Value']]

    # Scale the data (important for K-Means)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_cluster)

    # Determine optimal number of clusters (Elbow Method)
    inertia = []
    k_range = range(2, 6) # Try k=2 to 5 as requested
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # Added n_init for reproducibility
        kmeans.fit(X_scaled)
        inertia.append(kmeans.inertia_)

    # Plot Elbow Method
    plt.figure(figsize=(8, 5))
    plt.plot(k_range, inertia, marker='o')
    plt.title('Elbow Method for Optimal K')
    plt.xlabel('Number of Clusters (K)')
    plt.ylabel('Inertia')
    plt.xticks(k_range)
    plt.grid(True)
    plt.savefig("kmeans_elbow_method.png", dpi=300, bbox_inches="tight")
    plt.show()
    print("Elbow method plot saved as kmeans_elbow_method.png")

    # Choose a K based on the elbow plot (e.g., let's choose k=3 for demonstration if the elbow is not clear)
    # You might need to adjust this based on the actual plot
    optimal_k = 3 # Example: Choose 3 clusters

    print(f"\nUsing K = {optimal_k} for K-Means clustering.")

    # Apply K-Means with the chosen K
    kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
    df_merged['cluster'] = kmeans.fit_predict(X_scaled)

    # Add scaled features back to the merged dataframe for visualization
    df_merged['co2_scaled'] = X_scaled[:, 0]
    df_merged['aqi_scaled'] = X_scaled[:, 1]


    # Visualize K-Means Clusters using Plotly (Interactive)
    print("\nVisualizing K-Means Clusters...")
    fig_kmeans = px.scatter(df_merged,
                            x='co2_scaled',
                            y='aqi_scaled',
                            color='cluster',
                            hover_name='country_name_norm',
                            title=f'K-Means Clustering (K={optimal_k}): CO2 vs AQI (Scaled)',
                            labels={'co2_scaled': 'Scaled CO2 Emissions (kt)', 'aqi_scaled': 'Scaled Mean AQI Value'})
    fig_kmeans.write_html(f"kmeans_clusters_k{optimal_k}.html") # Save as HTML
    fig_kmeans.show()
    print(f"K-Means clustering scatter plot saved as kmeans_clusters_k{optimal_k}.html (HTML)")

    # Visualize K-Means Clusters using Matplotlib (Static)
    plt.figure(figsize=(10, 7))
    sns.scatterplot(data=df_merged, x='co2_scaled', y='aqi_scaled', hue='cluster', palette='viridis', legend='full')
    plt.title(f'K-Means Clustering (K={optimal_k}): CO2 vs AQI (Scaled)')
    plt.xlabel('Scaled CO2 Emissions (kt)')
    plt.ylabel('Scaled Mean AQI Value')
    plt.savefig(f"kmeans_clusters_k{optimal_k}.png", dpi=300, bbox_inches="tight")
    plt.show()
    print(f"K-Means clustering scatter plot saved as kmeans_clusters_k{optimal_k}.png (PNG)")

    # Display countries in each cluster
    print("\nCountries in Each K-Means Cluster:")
    for cluster_id in sorted(df_merged['cluster'].unique()):
        print(f"\n--- Cluster {cluster_id} ---")
        cluster_countries = df_merged[df_merged['cluster'] == cluster_id][['country_name_norm', 'co2_emission_kt_avg_2017_2019', 'AQI Value']].sort_values(by='co2_emission_kt_avg_2017_2019', ascending=False)
        display(cluster_countries)


    # Calculate mean values per cluster for interpretation
    print("\nMean CO2 and AQI Values per Cluster:")
    cluster_summary = df_merged.groupby('cluster')[['co2_emission_kt_avg_2017_2019', 'AQI Value']].mean()
    display(cluster_summary)


else:
    print("\nSkipping Clustering: Merged data not available or missing required columns.")


# ----------------------------------------------------------------------------------------------------
# 5) Optional: Hierarchical Clustering
# ----------------------------------------------------------------------------------------------------
if 'X_scaled' in locals() and X_scaled is not None:
    print("\n5. Optional: Hierarchical Clustering...")
    try:
        # Perform hierarchical clustering
        linked = linkage(X_scaled, method='ward') # Using Ward's method

        # Plot dendrogram
        plt.figure(figsize=(12, 8))
        dendrogram(linked,
                   orientation='top',
                   labels=df_merged['country_name_norm'].tolist(), # Use country names as labels
                   distance_sort='descending',
                   show_leaf_counts=True)
        plt.title('Hierarchical Clustering Dendrogram')
        plt.xlabel('Country')
        plt.ylabel('Distance')
        # Adjust layout to prevent labels overlapping if many countries
        plt.tight_layout()
        plt.savefig("hierarchical_clustering_dendrogram.png", dpi=300)
        plt.show()
        print("Hierarchical clustering dendrogram saved as hierarchical_clustering_dendrogram.png")

        print("\nHierarchical Clustering performed. Review the dendrogram to see potential groupings.")

    except Exception as e:
        print(f"\nError during Hierarchical Clustering: {e}")
        print("Skipping Hierarchical Clustering.")
else:
     print("\nSkipping Hierarchical Clustering: Scaled data for clustering not available.")


# ----------------------------------------------------------------------------------------------------
# 6) Summarize Findings & 7) Finish task
# ----------------------------------------------------------------------------------------------------
print("\n=== Advanced Analysis Completed ===")
print("Review the plots and country lists generated above to interpret the results.")
print("- **Forecasting**: Check the CO2 trend and the Linear Regression and ARIMA forecasts.")
print("- **Clustering**: Examine the Elbow Method plot to justify the chosen K. Analyze the scatter plots and the list of countries in each cluster to understand the country segments based on CO2 emissions and AQI Value.")
print("All generated plots have been saved as PNG/HTML files in the current directory.")
```
